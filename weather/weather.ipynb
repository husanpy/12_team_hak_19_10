```python
import numpy as np
import pandas as pd

import bq_helper 

bqh = bq_helper.BigQueryHelper(active_project= "bigquery-public-data", dataset_name= "noaa_gsod")

START_YEAR = 2000
END_YEAR = 2020

countries = ['TU', 'UA', 'EG', 'KZ', 'MX', 'ZA', 'AR', 'AU', 'BR', 'CA']
stn_query = "SELECT usaf AS stn, name, country, state, lat, lon, elev FROM `bigquery-public-data.noaa_gsod.stations` WHERE country IN ('TU', 'UA', 'EG', 'KZ', 'MX', 'ZA', 'AR', 'AU', 'BR', 'CA') ORDER BY stn"
df_stn = bqh.query_to_pandas_safe(stn_query, max_gb_scanned=5)
stn_list = ','.join(["'"+i+"'" for i in df_stn['stn']])
df_stn = df_stn.drop(['name', 'state', 'lat', 'lon', 'elev'], axis=1)
filename = 'station_country.csv'
df_stn.to_csv(filename, index = False)
print ("Saved {}".format(filename))



for year in range(START_YEAR, END_YEAR):
    query = "SELECT stn, year, AVG(temp) as temp, AVG(dewp) as dewp, AVG(slp) as slp, AVG(stp) as stp, AVG(visib) as visib, sum(prcp) as prcp FROM `bigquery-public-data.noaa_gsod.gsod{}` WHERE stn IN ({})GROUP BY stn, year ORDER BY stn".format(year, stn_list)
    df_wthr = bqh.query_to_pandas_safe(query, max_gb_scanned=5)
    filename = 'weather_{}.csv'.format(year)
    df_wthr.to_csv(filename, index = False)
    print ("Saved {}".format(filename))
    

```
