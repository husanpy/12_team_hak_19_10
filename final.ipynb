```python
import pandas as pd
```

```python
weather_weather = pd.read_csv(r'C:\Users\glll4\weather\full_weather.csv')
station_weather = pd.read_csv(r'C:\Users\glll4\station_country.csv')

weather = weather_weather.set_index('stn').join(station_weather.set_index('stn'))
weather = weather.groupby(by=['year','country']).mean().reset_index(0).reset_index()
weather.loc[weather['country'] == 'TU', 'country'] = 'Turkey'
weather.loc[weather['country'] == 'UA', 'country'] = 'Ukraine'
weather.loc[weather['country'] == 'AR', 'country'] = 'Argentina'
weather.loc[weather['country'] == 'AU', 'country'] = 'Australia'
weather.loc[weather['country'] == 'KZ', 'country'] = 'Kazakhstan'
weather.loc[weather['country'] == 'CA', 'country'] = 'Canada'
weather.loc[weather['country'] == 'MX', 'country'] = 'Mexico'
weather.loc[weather['country'] == 'ZA', 'country'] = 'South Africa'
weather.loc[weather['country'] == 'BR', 'country'] = 'Brazil'
weather.loc[weather['country'] == 'EG', 'country'] = 'Egypt'
weather['year'] = weather['year'].astype(str)
```

```python
yield_data = pd.read_csv(r'C:\Users\glll4\Downloads\aas.csv')
yield_data = yield_data.drop(['Data Source', 'Element Name'], axis=1)
yield_data.columns = ['country', 'product', 'year', 'yield']
for i in range(len(yield_data)):
    yield_data.year[i] = yield_data.year[i][0:4]
```

```python
final_df = pd.merge(yield_data.reset_index(), weather.reset_index(), on='country')
final_df = final_df[final_df.year_y == final_df.year_x].reset_index().drop(['index', 'index_x', 'year_y','index_y'], axis=1)
final_df.columns = ['country','product','year','yield','temperature','dew_point','sea_level_pressure','station_pressure','visibility','precipitation']
final_df = final_df[['country','year','product','yield','temperature','dew_point','sea_level_pressure','station_pressure','visibility','precipitation']]
final_df = final_df.drop(['dew_point','visibility'], axis=1)
final_df.temperature = 5 / 9 * (final_df.temperature - 32)
```

```python
final_df.to_csv('final_dataset.csv', index=False)  
```
